## QUESTION 1 - When is the best time of day, day of the week, and time of year to fly to minimise delays?

## Best time of the day

#### To minimize delay, we need to find the average delay and the total delayed flights each flights DepTime

```python
flight_filtered = flight[flight.ArrDelay>0]
flight_grouped = flight_filtered.groupby('DepTime').agg({'ArrDelay':['mean', 'count']}).reset_index()
flight_grouped.columns = ['DepTime','average_delay', 'delay_count']
Best_time=flight_grouped.sort_values(by=['average_delay'])
Best_time
```

### Visualize using line graph

```python
Best_time.plot.line(x='average_delay', y='DepTime', title='Best Time of The Day', color='#FAAFBE')
```

## Best day of the week

#### To minimize delay in the span of 7 days, we need to find the total flights and its average delay on each day of week

```python
#Best day of week
flight_filtered_1 = flight[flight.ArrDelay>0]
flight_grouped_1 = flight_filtered_1.groupby('DayOfWeek').agg({'ArrDelay':['mean', 'count']}).reset_index()
flight_grouped_1.columns = ['DayOfWeek','average_delay', 'delay_count']
Best_day=flight_grouped_1.sort_values(by=['average_delay'])
Best_day
```

### Visualize using horizontal bar chart

```python
Best_day.plot.barh(x='DayOfWeek', y='average_delay', title='Best Day of The Week', color = ['lavenderblush','pink', 'lightpink', 'hotpink', 'orchid', 'violet', 'plum'], figsize=(10,6))
```

## Best time of the year

#### To minimize delay in the span of 12 months, we need to find the total flights and its average delay on each month

```python
#best time of year
flight_filtered_2 = flight[flight.ArrDelay>0]
flight_grouped_2 = flight_filtered_2.groupby('Month').agg({'ArrDelay':['mean','count']}).reset_index()
flight_grouped_2.columns = ['Month','average_delay', 'delay_count']
Best_time_year=flight_grouped_2.sort_values(by=['average_delay'])
Best_time_year
```

### Visualize using horizontal bar chart

```python
Best_time_year.plot.barh(x='Month', y='average_delay', title='Best Time of The Year', rot = 0, figsize=(10,6), color='pink')
```

## QUESTION 5 - Use the available variables to construct a model that predicts delays

### Set a sample of 100,000 datas only from the flight data

```python
sample_flight = flight.sample(n=100000,replace="False")
```

### Checking whether the sample_flight have missing values

```python
sample_flight.isnull().sum().sum()
```

### Set some columns from sample_flight as the numerical values, denotes as X

```python
X = sample_flight[['Year', 'Month','DayofMonth','DepTime','CRSDepTime','ArrTime','CRSArrTime','DepDelay']]
```

### Set a column from sample_flight as the target value, denotes as y

```python
y = sample_flight['ArrDelay']
```

### Split the data into train and test set with ratio of 7:3

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

### Calculating MSE using linear regression

```python
model = LinearRegression().fit(X_train, y_train)
y_pred_ln_t = model.predict(X_train)
y_pred_ln = model.predict(X_test)

mse_train_2 = mean_squared_error(y_true=y_train, y_pred=y_pred_ln_t, squared=True)
mse_test_2 = mean_squared_error(y_true=y_test, y_pred=y_pred_ln, squared=True)
r_squared_2 = model.score(X_train, y_train)
print("MSE train :", mse_train_2)
print("MSE test :", mse_test_2)
print("R squared :", r_squared_2)
```

### Calculating MSE using ridge regression

```python
ridge_mod=Ridge(alpha=0, normalize=True).fit(X_train,y_train)
y_pred_rd_t = ridge_mod.predict(X_train)
y_pred_rd = ridge_mod.predict(X_test)

mse_train_1 = mean_squared_error(y_true=y_train,y_pred=y_pred_rd_t, squared=True)
mse_test_1 = mean_squared_error(y_true=y_test,y_pred=y_pred_rd, squared=True)
r_squared_1 = ridge_mod.score(X_train, y_train)
print("MSE train :", mse_train_1)
print("MSE test :", mse_test_1)
print("R squared :", r_squared_1)
```

### Calculating MSE using lasso regression

```python
lasso_regression = Lasso(alpha=1).fit(X_train,y_train)
lasso_predict_train = lasso_regression.predict(X_train)
lasso_predict_test = lasso_regression.predict(X_test)

mse_train = mean_squared_error(y_true=y_train, y_pred=lasso_predict_train, squared=True)
mse_test = mean_squared_error(y_true=y_test, y_pred=lasso_predict_test, squared=True)
r_squared = lasso_regression.score(X_train, y_train)
print("MSE train :", mse_train)
print("MSE test :", mse_test)
print("R squared :", r_squared)
```

### Calculating MSE using random forest

```python
model1 = RandomForestRegressor(min_samples_leaf=1).fit(X_train, y_train)
y_pred_rf_t = model1.predict(X_train)
y_pred_rf = model1.predict(X_test)
mse_train_3 = mean_squared_error(y_true=y_train, y_pred=y_pred_rf_t, squared=True)
mse_test_3 = mean_squared_error(y_true=y_test, y_pred=y_pred_rf, squared=True)
r_squared_3 = model1.score(X_train, y_train)
print("MSE train :", mse_train_3)
print("MSE test :", mse_test_3)
print("R squared :", r_squared_3) 
```
